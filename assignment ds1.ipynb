```python
# importing package
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#Reading csv data
filepath="/Users/arunmergu/Downloads/Cars_Multi.csv"

def read_csv(filepath):
    """ It will load the csv data from the given file path
    """
    cars = pd.read_csv(filepath)
    return cars

cars=read_csv(filepath)
y=cars['origin']
x=cars['cylinders']
cars
```

```python
#Method1
def linearplt(x,y):
    """it shows the line plot between origin and cylinder"""
    plt.plot(list(x),list(y), label='line plot between origin and cylinders')
    plt.title('')
    plt.xlabel('cylinders')
    plt.ylabel('origin')
    plt.legend(loc='upper right')
    plt.grid()
    plt.show()

linearplt(x,y)
```

```python
#Method2

def histplt(data):
    """histogram of a variable in cars data"""
    plt.figure(figsize=(25, 6))
    plt.subplot(1,3,2)
    plt1 = cars.cylinders.value_counts().plot(kind='bar')
    plt.title('Number of cyliners Histogram')
    plt1.set(xlabel = 'cylinders', ylabel='Frequency of cylinders')

histplt(cars)
```

```python
#Method3
def scatter(x,fig):
    
    """it shows the scatter plot between the acceleration and horsepwer"""
    plt.subplot(5,2,fig)
    plt.scatter(cars[x],cars['horsepower'])
    plt.title(x+' vs horsepower')
    plt.ylabel('horsepower')
    plt.xlabel(x)
    

plt.figure(figsize=(10,20))

scatter('acceleration', 1)
```
